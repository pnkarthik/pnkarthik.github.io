# Sorting

```python
colors = ["red", "yellow", "green", "blue"]
colors.sort()
colors
```

```python
numbers = [1, 10, 5, 3]
numbers.sort()
numbers
```

```python
colors = ["red", "yellow", "green", "blue"]
colors.sort(key=len)
colors
```

```python
# any user defined function can be passed to key, but the function should take
# as input only one single argument.

def get_second_element(item):
    return item[1]

items = [(4, 1), (1, 2), (-9, 0)]
items.sort(key=get_second_element)
items
```

```python
items = [(4, 1), (1, 2), (-9, 0)]
items.sort()
items
```

```python
items = ["word", 1, (-9, 0)]
items.sort()
items
```
