## The `While` Loop

```python
n = 1
while n < 5:
    print(n)
    n = n + 1
```

```python
# while loop serves its purpose as long as a block of code needs to be executed until a certain condition is met.
# if a loop must run for a certain specific number of times, a better alternative is the 'for' loop.
```

## Scope of Variables in Python

```python
x = "Hello World"

def func():
    x = 2
    print(f"Inside 'func', x has the value {x}")

func()
print(f"Outside 'func', x has the value {x}")
```

```python
x = 5
def outer_func():
    y = 3
    def inner_func():
        z = x + y
        return z
    return inner_func()

print(outer_func())
print(type(outer_func()))
```

```python
total = 10
def add_to_total(n, total):
    total = total + n
    print(f"Value of total within function is {total}")

add_to_total(5, total)
print(f"Value of total within function is {total}")
```
