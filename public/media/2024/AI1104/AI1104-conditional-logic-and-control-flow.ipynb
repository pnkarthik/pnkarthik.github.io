## The `if-elif-else` Statement

```python
grade = 85  #1

if grade >= 90:  #2
    print("You passed the class with a A.")
elif grade >= 80:  #3
    print("You passed the class with a B.")
elif grade >= 70:  #4
    print("You passed the class with a C.")
else:  #5
    print("You did not pass the class :(")

print("Thanks for attending.")  #6
```

```python
sport = input("Enter a sport: ")
p1_score = int(input("Enter player 1 score: "))
p2_score = int(input("Enter player 2 score: "))

# 1

if sport.lower() == "basketball":
    if p1_score == p2_score:
        print("The game is a draw.")
    elif p1_score > p2_score:
        print("Player 1 wins.")
    else:
        print("Player 2 wins.")
# 2
elif sport.lower() == "golf":
    if p1_score == p2_score:
        print("The game is a draw.")
    elif p1_score < p2_score:
        print("Player 1 wins.")
    else:
        print("Player 2 wins.")
# 3
else:
    print("Unknown sport")
```

```python
# a simplification

if p1_score == p2_score:
    print("The game is a draw.")

elif sport.lower() == "basketball":
    if p1_score > p2_score:
        print("Player 1 wins.")
    else:
        print("Player 2 wins.")
elif sport.lower() == "golf":
    if p1_score < p2_score:
        print("Player 1 wins.")
    else:
        print("Player 2 wins.")
else:
    print("Unknown sport.")
```

## Combining `if` Statement with `For` Loops

```python
# sum of all even numbers up to 100

sum_of_evens = 0

for n in range(1, 101):
    if n % 2 == 0:
        sum_of_evens = sum_of_evens + n

print(sum_of_evens)
```

```python
for n in range(0, 4):
    if n == 2:
        break
    print(n)

print(f"Finished with n = {n}")
```

```python
for n in range(0, 4):
    if n == 2:
        continue
    print(n)

print(f"Finished with n = {n}")
```

## Combining `for` and `else`

```python
# look for 'x' in a string

phrase = "it marks the spot"

for character in phrase:
    if character == "X":
        break
else:
    print("No 'x' was found in the phrase")
```

Remark: `break`, `continue`, and `else` can be used along with `while` loops also.

## `try` and `except`

```python
while(1):
    try:
        number = int(input("Enter an integer: "))
        break
    except ValueError:
        print("That was not an integer")

print (f"The entered integer was {number}")
```

```python
def divide(num1, num2):
    '''This is a docstring for divide function'''
    try:
        print(num1 / num2)
    except TypeError:
        print("Both arguments must be numbers")
    except ZeroDivisionError:
        print("num2 must not be 0")

divide
```
