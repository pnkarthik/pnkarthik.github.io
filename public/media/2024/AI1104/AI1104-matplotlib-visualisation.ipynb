```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
np.random.seed(444)  # same reproducible (pseudo)random data
```

```python
rng = np.arange(50)
print(rng)
```

```python
# np.random.seed(444)  # same reproducible (pseudo)random data
rnd = np.random.randint(0, 10, size=(3, rng.size))
print(rnd)
```

```python
yrs = 1950 + rng
```

```python
print(rnd + rng)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.stackplot(yrs, rng+rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_label('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])

plt.show()
```

```python
x_data = np.random.randint(1, 11, size=50)
```

```python
y_data = x_data + np.random.randint(1, 4, size=x_data.size)
```

```python
data = np.column_stack([x_data, y_data])
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax1.scatter(x=x_data, y=y_data, marker='o', color='red', edgecolor='blue')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_axisbelow(True)  # apply ticks and grid 'behind' the points
ax1.grid(linestyle='--')

ax2.hist(data, bins=np.arange(data.min(), data.max()), label=('$x$', '$y$'))
ax2.legend(loc=(0))  # render legend in the best possible location
ax2.set_title("Frequencies of $x$ and $y$")
ax2.yaxis.tick_right()  # the y-axis for the plot will be displayed on the right side of the plot

plt.show()
```
